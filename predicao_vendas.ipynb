```python
novo_df = {
    "nfe": [np.nan] * 3431,
    "vendedor_id": [np.nan] * 3431,
    "cliente_id": [np.nan] * 3431,
    "produto_id": [np.nan] * 3431,
    "valor_desconto": [np.nan] * 3431,
    "item_quantidade": [np.nan] * 3431,
    "valor_unitario": [np.nan] * 3431,
}
```

```python
df.nfe.unique()
```

```python
df_2 = pd.DataFrame(novo_df)
```

```python
df_2
```

```python
df_2.nfe = df.nfe.unique()
```

```python
df_2
```

```python
for i in range(len(df_2)):
    if df_2.nfe[i] == df.nfe.unique()[i]:
        df_2.vendedor_id[i] = df.vendedor_id[i]
```

```python
for i in range(len(df_2)):
    if df_2.nfe[i] == df.nfe.unique()[i]:
        df_2.cliente_id[i] = df.cliente_id[i]
```

```python
for i in range(len(df_2)):
    if df_2.nfe[i] == df.nfe.unique()[i]:
        df_2.produto_id[i] = df.produto_id[i]
```

```python
for i in range(len(df_2)):
    if df_2.nfe[i] == df.nfe.unique()[i]:
        df_2.valor_desconto[i] = df[df.nfe == df.nfe.unique()[i]][
            "valor_desconto"
        ].sum()
```

```python
for i in range(len(df_2)):
    if df_2.nfe[i] == df.nfe.unique()[i]:
        df_2.item_quantidade[i] = df[df.nfe == df.nfe.unique()[i]][
            "item_quantidade"
        ].sum()
```

```python
for i in range(len(df_2)):
    if df_2.nfe[i] == df.nfe.unique()[i]:
        df_2.valor_unitario[i] = df[df.nfe == df.nfe.unique()[i]][
            "valor_unitario"
        ].sum()
```

```python
df_2
```

```python
df = df_2.copy()
```

```python
df
```

```python
df["valor_total"] = df["valor_unitario"] - df["valor_desconto"]
```

```python
df
```

```python
df.cliente_id = df.cliente_id.apply(lambda x: int(x))
df.produto_id = df.produto_id.apply(lambda x: int(x))
df.item_quantidade = df.item_quantidade.apply(lambda x: int(x))
```

```python
df
```

```python
df.to_csv("data_cleaned.csv", sep=",", index=False)
```

```python
X = df.drop(columns=["nfe", "valor_desconto", "valor_unitario", "valor_total"])
y = df["valor_total"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
depth = 5
```

```python
model = DecisionTreeRegressor(max_depth=depth)
model = model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
df_temp = pd.DataFrame({"Real": y_test, "Predict": y_pred})
df_temp.sample(5)

display(df_temp.sample(5))
```

```python
acc = model.score(x_test, y_test)
print(f"Model Accuracy = {acc:.2%}")
```
